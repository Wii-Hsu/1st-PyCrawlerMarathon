### 根據範例檔案的結果：

1. 取出班次一的每一個時間
2. 將班次一的每一個時間用一種資料型態保存
3. 將班次一到五與其所有時間用一種資料型態個別保存

```python
# 1. 取出班次一的每一個時間
import csv

with open('example.csv','r',encoding="utf-8",newline='') as csvfp:
 rows = csv.reader(csvfp)

 for x in rows:
    print(x[0],"\t",x[5])
 """       
    #序號
    print(x[0],"\t",end='') 
    for i in range(5,16):
        if i==5:
            print(x[i],end='')
         #最後一個換行   
        elif i==15:
            print("\t",x[i])
        else:
             print("\t",x[i],end='')
"""
```

```python
# 2. 將班次一的每一個時間用一種資料型態保存

import csv,json

jsonFilePath = "D2_1.json"

data=[]
with open('example.csv','r',encoding="utf-8",newline='') as csvfp:
 rows = csv.reader(csvfp)

 for x in rows:
    data.append(x[5])
 print(data)

with open(jsonFilePath,'w',encoding='utf8') as jsonfp:
    jsonfp.write(json.dumps(data, indent = 4,ensure_ascii=False))

    
with open(jsonFilePath,'r',encoding="utf-8") as jsonfp1:
    print(jsonfp1.read())
```

```python
# 3. 將班次一到五與其所有時間用一種資料型態個別保存

import csv,json

jsonFilePath = "D2_2.json"

data=[]
with open('example.csv','r',encoding="utf-8",newline='') as csvfp:
 rows = csv.reader(csvfp)

 for x in rows:
    data.append(x)
 print(data)

with open(jsonFilePath,'w',encoding='utf8') as jsonfp:
    jsonfp.write(json.dumps(data, indent = 4,ensure_ascii=False))

    
with open(jsonFilePath,'r',encoding="utf-8") as jsonfp1:
    print(jsonfp1.read())
```
